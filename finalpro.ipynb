```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
orders_df = pd.read_csv('Orders.csv')
details_df = pd.read_csv('Details.csv')

# Merge the datasets on 'Order ID'
merged_df = pd.merge(orders_df, details_df, on='Order ID')

# Convert 'Order Date' to datetime format
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'], format='%d-%m-%Y')

# Calculate total profit and amount generated per day
daily_summary = merged_df.groupby('Order Date').agg({'Profit': 'sum', 'Amount': 'sum'}).reset_index()

# Rename columns for clarity
daily_summary.columns = ['Order Date', 'Total Profit', 'Total Amount']

# Display the summary
print(daily_summary)

# Plotting the results
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_summary, x='Order Date', y='Total Profit', label='Total Profit', marker='o')
sns.lineplot(data=daily_summary, x='Order Date', y='Total Amount', label='Total Amount', marker='o')
plt.title('Total Profit and Amount Generated per Day')
plt.xlabel('Order Date')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
orders_df = pd.read_csv('Orders.csv')
details_df = pd.read_csv('Details.csv')

# Merge the datasets on 'Order ID'
merged_df = pd.merge(orders_df, details_df, on='Order ID')

# Calculate total quantity, profit, and amount by category
category_summary = merged_df.groupby('Category').agg({
    'Quantity': 'sum',
    'Profit': 'sum',
    'Amount': 'sum'
}).reset_index()

# Rename columns for clarity
category_summary.columns = ['Category', 'Total Quantity', 'Total Profit', 'Total Amount']

# Display the summary
print(category_summary)

# Plotting the results
plt.figure(figsize=(12, 6))
sns.barplot(data=category_summary, x='Category', y='Total Quantity', color='blue', alpha=0.6, label='Total Quantity')
sns.barplot(data=category_summary, x='Category', y='Total Profit', color='green', alpha=0.6, label='Total Profit')
sns.barplot(data=category_summary, x='Category', y='Total Amount', color='orange', alpha=0.6, label='Total Amount')

plt.title('Total Quantity, Profit, and Amount by Category')
plt.xlabel('Category')
plt.ylabel('Total')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
